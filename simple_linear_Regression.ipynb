```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel('/content/drive/MyDrive/ML Learning/car_sales_data.xlsx')
df.head()
```

```python
plt.xlabel('Engine_Size', color='red')
plt.ylabel('Price', color='red')
plt.scatter(df.Engine_Size, df.Price, color='blue', marker='+')
```

Model Training

```python
# Object Creation
lr = linear_model.LinearRegression()
lr.fit(df[['Engine_Size']], df.Price)
```

Prediction Test

```python
lr.predict([[4.5]])
```

```python
lr.score(df[['Engine_Size']], df.Price)
```

```python
lr.coef_  # slope
```

```python
lr.intercept_
```
